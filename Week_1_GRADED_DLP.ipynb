```python
def count_tokens(example):
    # 'tokens' is the existing column in the Naamapadam dataset
    example['num_tokens'] = len(example['tokens'])
    return example

# Apply the map function to all splits (train, test, validation)
ds = ds.map(count_tokens)

# Calculate total tokens across all splits
total_tokens = sum(sum(split['num_tokens']) for split in ds.values())

# Convert to millions
tokens_in_millions = total_tokens / 1_000_000

print(f"Total tokens: {tokens_in_millions:.2f} million")
```

```python
from datasets import concatenate_datasets

# 1. Concatenate the splits in the specific order: train, test, validation
combined_ds = concatenate_datasets([ds['train'], ds['test'], ds['validation']])

# 2. Define a function to join tokens into a single string
def join_tokens(example):
    example['text'] = " ".join(example['tokens'])
    return example

# 3. Apply the function to create the 'text' column
#    (We process the combined dataset here)
ds_mapped = combined_ds.map(join_tokens)

# 4. Remove the old columns and reassign to 'ds'
ds = ds_mapped.remove_columns(['ner_tags', 'tokens'])

# 5. Print the total number of samples
print(f"Total number of samples in ds: {len(ds)}")
```

```python
# Define the filter function
def filter_min_tokens(example):
    # Split by whitespace and count tokens
    return len(example['text'].split()) >= 6

# Apply the filter to the dataset
ds = ds.filter(filter_min_tokens)

# Print the number of samples
print(len(ds))
```

```python
from datasets import load_dataset, concatenate_datasets, interleave_datasets
import numpy as np

# 1. Process Naamapadam (Tamil)
# Load the dataset
ds_naam = load_dataset("ai4bharat/naamapadam", "ta")

# Concatenate all splits [train, test, validation]
ds_naam_concat = concatenate_datasets([ds_naam['train'], ds_naam['test'], ds_naam['validation']])

# Create 'text' column from 'tokens' and remove old columns
def process_naam(example):
    example['text'] = " ".join(example['tokens'])
    return example

ds_naam_processed = ds_naam_concat.map(process_naam, remove_columns=['tokens', 'ner_tags'])

# Filter: at least 6 words
# Note: "text" is space-separated words
ds_naam_filtered = ds_naam_processed.filter(lambda x: len(x['text'].split()) >= 6)


# 2. Process IndicGlue (Tamil Headlines)
# Load the dataset
ds_glue = load_dataset("ai4bharat/indic_glue", "inltkh.ta")

# Concatenate all splits (train, validation, test)
# Note: We convert values to a list to handle the DatasetDict
ds_glue_concat = concatenate_datasets(list(ds_glue.values()))

# Filter: at least 6 words
# Note: IndicGlue already has a 'text' column
ds_glue_filtered = ds_glue_concat.filter(lambda x: len(x['text'].split()) >= 6)


# 3. Interleave Datasets
# "Take 80% of samples from naamapadam and 20% from indic glue"
# We align the order: [Naamapadam, IndicGlue] -> [0.8, 0.2]
# Stopping strategy defaults to "first_exhausted" which limits size based on the smaller dataset
ds_interleaved = interleave_datasets(
    [ds_naam_filtered, ds_glue_filtered],
    probabilities=[0.8, 0.2],
    seed=42,
    stopping_strategy="first_exhausted"
)

# 4. Result
print(f"Number of samples after interleaving: {len(ds_interleaved)}")
```
