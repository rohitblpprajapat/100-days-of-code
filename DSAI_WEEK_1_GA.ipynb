```python
t = np.linspace(1, 5, 10000)

# Define the left-hand side (LHS) and right-hand side (RHS)
lhs = t * np.sin(t)
rhs = np.exp(-t) / (t**2)

# Find the index where the absolute difference is smallest
difference = np.abs(lhs - rhs)
min_index = np.argmin(difference)
solution = t[min_index]

print(f"Q6 Answer - Solution for x: {solution:.2f}")

# Plotting the graph (as requested by the problem statement)
plt.figure(figsize=(10, 6))
plt.plot(t, lhs, )
plt.plot(t, rhs, )
plt.axvline(x=solution, color='r', linestyle='--')
plt.title('Graphical Solution for Q6')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
from tensorflow.keras.datasets import mnist

# ==========================================
# PART 1: QUESTIONS 8 & 9 (MNIST)
# ==========================================
print("--- Questions 8 & 9 ---")

# Load the MNIST dataset
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# --- Question 8 ---
# Find the number of images with class label 5
num_fives = np.sum(y_train == 5)
print(f"Q8 Answer: Number of images with label 5: {num_fives}")

# --- Question 9 ---
# Find number of images with class label 7 where pixel (14, 14) is non-zero
# Step 1: Filter for images with label 7
indices_7 = np.where(y_train == 7)[0]
images_7 = X_train[indices_7]

# Step 2: Extract the pixel at (14, 14) for these images
# The shape is (N, 28, 28), so we access [:, 14, 14]
pixels_14_14 = images_7[:, 14, 14]

# Step 3: Count non-zero entries
count_nonzero = np.count_nonzero(pixels_14_14)
print(f"Q9 Answer: Number of 7s with non-zero pixel at (14,14): {count_nonzero}")
print("-" * 30)

# ==========================================
# PART 2: QUESTION 10 (Matrices)
# ==========================================
print("--- Question 10 ---")

# Setup from the image
rng = np.random.default_rng(seed=42)
A = rng.integers(0, 100, (100, 200))
B = rng.integers(0, 100, (100, 300))

# Concatenate them
# A is (100, 200), B is (100, 300).
# They share the first dimension (rows), so we concatenate along axis 1 (columns).
C = np.concatenate((A, B), axis=1)

# Find the mean of the elements in the 10th row
# "10th row" means index 9 (since the problem says 'top-most row is the first')
row_10 = C[9, :]
mean_val = np.mean(row_10)

print(f"Q10 Answer: Mean of the 10th row: {mean_val:.2f}")
```
