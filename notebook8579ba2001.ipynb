```python
# 15 make a python function that handles scalars to work on numpy arrays


def maxx(x,y):
  return (x if x > y else y)

def pair_max(x,y):
  maximum = [maxx(a,b) for a,b in map(lambda a,b:(a,b), x,y)]
  return np.array(maximum)

a = np.array([5, 7, 9, 8, 6, 4, 5])
b = np.array([6, 3, 4, 8, 9, 7, 1])

pair_max(a,b)
```

```python
# 28 compute the mean, mode, median and standard deviation of a np array
import pandas as pd

housing_data = np.genfromtxt('/content/sample_data/california_housing_train.csv',
                             delimiter=',', usecols=[-1], skip_header=1,dtype=float)
# mean
print("mean",np.mean( housing_data))

# median
print('medeian', np.median(housing_data))

# mode
value, counts = np.unique(housing_data, return_counts=True)
max_value_index = np.argmax(counts)
print("mode: ", value[max_value_index])

# sd
print("SD", np.std(housing_data))

```
